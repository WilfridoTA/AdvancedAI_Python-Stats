## 1) General Customer Profile

Calculate the descrptive statistics for the **tenure**, **Monthly Charges**, and **Total charges** columns. By comparing the **mean** and **median** of **tenure**, how would you describe the company's "typical" customer?

```python
data.describe().T
```

```python
print("Moda")
print(f"tenure: {data['tenure'].mode()}")
print(f"Monthly charges: {data['MonthlyCharges'].mode()}")
print(f"Total Charges: {data['TotalCharges'].mode()}")
```

```python
plt.figure(figsize=(10,6))
plt.bar(['Tenure', 'MonthlyCharges', 'TotalCharges'], 
        [data['tenure'].mean(), data['MonthlyCharges'].mean(), data['TotalCharges'].mean()], 
        color=['blue', 'orange', 'green'])
plt.ylabel('Values')
plt.title('Mean Values of Tenure, Monthly Charges, and Total Charges')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(['Tenure', 'MonthlyCharges', 'TotalCharges'], 
        [data['tenure'].median(), data['MonthlyCharges'].median(), data['TotalCharges'].median()], 
        color=['blue', 'orange', 'green'])
plt.ylabel('Values')
plt.title('Median Values of Tenure, Monthly Charges, and Total Charges')
plt.grid(True)
plt.show()
```

Monthly charge -> Cuánto le están cobrando al cliente ahorita mismo

Total charges -> Cuánto se ha pagado a lo largo de su estadía

Conforme más tiempo se mantienen los clientes en la compañia aumenta su gasto total.

## 2) Distribution visualization

Generate **histograms** for the **tenure** and **Monthly Charges** columns. Describe the shape of each distribution. Are they symmetric, skewed? What interesting patters or peaks do you observe?

```python
plt.figure(figsize=(10,6))
plt.hist(data['tenure'], bins=30, edgecolor='white')
plt.title('Distribución de Tenure')
plt.xlabel('Tenure (meses)')
plt.ylabel('Número de Clientes')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(data['MonthlyCharges'], bins=30, edgecolor='white')
plt.title('Distribución de Monthly Charges')
plt.xlabel('Monthly Charges ($)')
plt.ylabel('Número de Clientes')
plt.show()
```

Mediana de tenure: 38 personas se quedan y 10 no se quedan
Mediana de monthly charges; 64.45 se quedan y 79.65 no se quedan
